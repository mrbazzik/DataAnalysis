### 2. Стабилизация дисперсии

Поскольку дисперсия меняется со временем, выполним преобразование Бокса-Кокса.

```python
from scipy import stats
data['wage_box'], lmbda = stats.boxcox(data.WAG_C_M)
data.wage_box.plot()
```

Проверим на стационарность

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_box)[1])
```

Гипотеза о нестационарности не отвергается, это подтверждается визуально видимым трендом.

### 3. Стационарность

Добьемся того, чтобы ряд стал стационарным. Сначала проведем сезонное дифференцирование.

```python
data['wage_box_sdiff'] = data.wage_box - data.wage_box.shift(12)
```

```python
sm.tsa.seasonal_decompose(data.wage_box_sdiff[12:]).plot()
```

Тренд уже не так ярко выражен, но по-прежнему присутствует.

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_box_sdiff[12:])[1])
```

Гипотеза о нестационарности отвергается на уровне 0.05. Несмотря на это, проведем еще обычное дифференцирование, чтобы полностью избавиться от тренда.

```python
data['wage_box_diff'] = data.wage_box_sdiff - data.wage_box_sdiff.shift(1)
sm.tsa.seasonal_decompose(data.wage_box_diff[13:]).plot()
```

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(data.wage_box_diff[13:])[1])
```

Гипотеза о нестационарности отвергается на уровне 0.01. Визуально тренда больше не видно.

### 4. Построение модели

Построим автокорреляцию итогового ряда

```python
sm.graphics.tsa.plot_acf(data.wage_box_diff[13:].values.squeeze(), lags=48)
```

Значимых сезонных лагов нет: Q = 0
В качестве q возьмем q = 5, посольку 5-й лаг чуть выше границы значимости.

Теперь посмотрим на частичную автокорреляцию

```python
sm.graphics.tsa.plot_pacf(data.wage_box_diff[13:].values.squeeze(), lags=48)
```

48-й лаг на границе значимости, возьмем P = 4
Посольку p должно быть меньше S (12), возьмем p = 10.

Сформируем списки параметров и выберем лучшую модель по метрике AIC.

```python
from itertools import product
ps = range(0, 11)
d=1
qs = range(0, 6)
Ps = range(0, 5)
D=1
Q = 0

parameters = product(ps, qs, Ps)
parameters_list = list(parameters)
len(parameters_list)
```

```python
import warnings
import numpy as np
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(data.wage_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, Q, 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except (ValueError, np.linalg.linalg.LinAlgError):
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Посмотрим на лучшую модель

```python
print(best_model.summary())
```

По критерию Льюнга-Бокса нестационарность остатков отвергается.

### 5. Анализ остатков

```python
best_model.resid[13:].plot()
```

Проверим смещенность остатков

```python
print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
```

Нулевая гипотеза о несмещенности остатков не отвергается.

Проверим стационарность остатков

```python
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Нулевая гипотеза о нестационарности отвергается.

```python
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48)
```

Неавтокоррелированность подтверждается визуально и по критерию Льюнга-Бокса из таблицы выше.

Отобразим предсказываемые значения на одном графике с наблюдаемыми.

```python
def invboxcox(y,lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
data['model'] = invboxcox(best_model.fittedvalues, lmbda)
data.WAG_C_M.plot()
data.model[13:].plot(color='r')
```

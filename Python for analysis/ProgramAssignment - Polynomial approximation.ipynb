```python
from math import sin,exp, pow
def f(x):
    return sin(x/5.)*exp(x/10.)+5*exp(-x/2.)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,17,0.1)
plt.plot(x, map(f,x))
```

```python
from scipy.linalg import solve
```

```python
a = np.array([[1,1],[1,15]])
b = [f(1), f(15)]
w = solve(a,b)
```

```python
def app(w, x):
    sum = w[0]
    for i in xrange(1,len(w)):
        sum+=w[i]*pow(x, i)
    return sum
```

```python
plt.plot(x, map(f,x))
plt.plot(x, map(lambda x1: app(w, x1), x))
```

```python
a = np.array([[1,1,1],[1,8,64],[1,15,225]])
b = [f(1),f(8),f(15)]
w = solve(a, b)
```

```python
plt.plot(x, map(f,x))
plt.plot(x, map(lambda x1: app(w, x1), x))
```

# Прогнозирование оттока клиентов

### <u>Цель проекта</u>

Выстраивание взаимоотношений с клиентами или customer relationship managmenet (CRM) является важным звеном в функционировании любого бизнеса. От того, насколько хорошо простроена работа с аудиторией зачастую зависит успех всей компании. Одна из наиболее актуальных задач из области CRM: прогнозирование оттока пользователей или churn prediction. Суть задачи заключается в заблаговременном нахождении сегмента пользователей, склонных через некоторый промежуток времени отказаться от использования некоторого продукта или услуги. Точное и своевременное нахождение таких пользователей позволяет эффективно бороться с их оттоком, например, выявлять причины оттока и принимать меры по удержанию клиентов. Эта задача актуальна для большинства организаций, оказывающих услуги в сегменте B2C и вдвойне актуальна в областях, где распространение услуги близко к отметке 100%. В такой ситуации задача удержания клиентов и выстраивания с ними взаимоотношений выходит на первый план.

### <u>Задачи проекта</u>

1\. Математически поставить задачу прогнозирования оттока клиентов.<br>
2\. Построить и оптимизировать прогнозную модель.<br>
3\. Оценить качество модели и экономический потенциал.

### <u>Описание технического решения</u>

- Данные содержат большое количество пропусков. Из модели полностью удалены признаки, содержащие более 50% пропущенных значений.
- Оставшиеся пропуски в чиловых переменных заменены на моды признаков.
- Категориальные признаки обработаны посредством dummy-кодирования. В случаях, когда категориальный признак имел слишком много категорий, но при этом сильно коррелировал с целевой переменной, использовался LabelEncoder.
- К данным применен undersampling по методу NearMiss.
- В качестве модели используется логистическая регрессия с l1-регуляризацией и c=0.5.
- В качестве метрики качества используется PR AUC, дополнительно используется ROC AUC и Precision.

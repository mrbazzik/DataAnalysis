3\. Рассчитайте корреляции переменных с целевой функцией и проанализируйте полученные данные.

Обратите внимание на то, что целевая функция предсталвляет собой бинарную переменную. Теоретически, это не помешает рассчиать, например, корреляцию Пирсона между непрерывной переменной и целевой функцией, онднако анализиоовать полученные результаты будет сложно, корреляция Пирсона просто не рассчитана на такое применение. Кто уже забыл, как действовать в такой ситуации - вспоминить можно вот здесь: https://www.coursera.org/teach/stats-for-data-analysis/content/edit/lecture/prX3S/video-subtitles 

Разделим признаки на числовые и категориальные

```python
real_features = ["Var"+str(i) for i in xrange(1, 191)]
cat_features = ["Var"+str(i) for i in xrange(191, 231)]
```

Рассчитаем корреляцию между вещественными признаками и целевой переменной. Поскольку целевая переменная бинарная, в качестве оценки корреляции будем использовать разницу в средних, разделив значения признака в соответствии с значением целевой переменной (разницу посчитаем относительную, чтобы не завязываться на масштаб данных). Дополнительно оценим значимость этой разницы с помощью непараметрического критерия Манна-Уитни (чтобы не анализировать распределения).

```python
from scipy.stats import mannwhitneyu
import numpy as np
diffs = []
for feat in real_features:
    data_yes = data.loc[labels[0]==1, feat]
    data_no = data.loc[labels[0]==-1, feat]
    mean_data_yes = data_yes.mean(skipna=True) 
    mean_data_no = data_no.mean(skipna=True)
    if ~np.isnan(mean_data_yes) and ~np.isnan(mean_data_no):
        temp = (feat, (mean_data_yes - mean_data_no)/float(mean_data_no), mannwhitneyu(data_yes, data_no).pvalue)
        diffs.append(temp)
```

Проанализируем только значимые разницы на уровне доверия 0.05

```python
diffs_filtered = [x for x in diffs if x[2]<0.05]
len(diffs_filtered)
```

Получили много достаточно много признаков со значимой разницей, выведем топ-10.

```python
diffs_filtered_sorted = sorted(diffs_filtered, key=lambda x: abs(x[1]), reverse=True)
diffs_filtered_sorted[:10]
```

Как видно, есть вещественные переменные с достаточно большой относительной разницей между средними.

Тепер рассмотрим категориальные признаки. Для оценки корреляции будем использовать коэффициент Крамера, значимость будем проверять с помощью критерия хи-квадрат.

```python
from scipy.stats import chi2_contingency
data["label"] = labels[0].values
coefs = []
for feat in cat_features:
    pt = data.pivot_table(index=feat, columns="label", aggfunc=len, values=data.index, fill_value=0)
    if pt.shape[0]>1 and pt.shape[1]>1:
        chi2_summary = chi2_contingency(pt) 
        n = pt.values.sum()
        kram_coef = np.sqrt(float(chi2_summary[0])/(n*min(pt.shape)-1))
        coefs.append((feat, kram_coef, chi2_summary[1]))
```

Проанализируем только значимые коэффициенты на уровне 0.05

```python
coefs_filtered = [x for x in coefs if x[2]<0.05]
len(coefs_filtered)
```

```python
coefs_filtered_sorted = sorted(coefs_filtered, key=lambda x: x[1], reverse=True)
coefs_filtered_sorted
```

По данным видно, что большинство категориальных признаков не имеют сильной связи с целевой переменной.

4\. Визуализируйте данные: рассмотрите топ 20 числовых переменных, наиболее сильно коррелирующих с целевой функцией. Для этих переменных постройте:

* Отобразите объекты в координатах пар признаков

Построим распределения в разрезе классов

```python
import matplotlib.pyplot as plt
import seaborn as sns
from math import *
% matplotlib inline
def plot_hists(data, features):
    n_cols = 5
    n_rows = ceil(len(features)/5.)
    fig = plt.figure(figsize=(3*n_cols, 3*n_rows))
    for i, feat in enumerate(features):
        ax = fig.add_subplot(n_rows,n_cols,i+1)
        data.loc[data.label==-1, feat].plot(kind='hist', ax=ax, color='blue', alpha=0.6, normed=True)
        data.loc[data.label==1, feat].plot(kind='hist', ax=ax, color='red', alpha=0.4, normed=True)

features = [x[0] for x in diffs_filtered_sorted[:20]]

plot_hists(data, features)
```

По визуальному анализу видно, что красный график (ушедшие клиенты) часто имеет бОльший разброс значений по признакам.

Отобразим объекты в разрезе пар признаков

```python
def plot_scatters(data, features):
    n_cols = 5
    n_rows = ceil(pow(len(features), 2)/5.)
    fig = plt.figure(figsize=(3*n_cols, 3*n_rows))
    j=1
    for i in xrange(len(features)):
        for k in xrange(i+1, len(features)):
            ax = fig.add_subplot(n_rows, n_cols, j)
            j+=1
            data.plot.scatter(features[i], features[k], c='label', ax=ax)

plot_scatters(data, features)
```

В результате визуального анализа в данном случае не видно признаков, по которым объекты очевидно разделимы.
